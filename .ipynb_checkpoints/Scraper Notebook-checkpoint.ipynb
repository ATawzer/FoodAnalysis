```python
import pandas as pd
from recipe_scrapers import scrape_me
import time
import random
import json
```

```python
def wait():
    
    x = random.randrange(2, 31, 1)
    print(f"Waiting for {x} Seconds.", end='\r')
    time.sleep(x)

def scrape_database():
    
    # read in database
    path = 'recipe_url_database.csv'
    output = 'scraped_database.csv'    
    df = pd.read_csv(path)
    df_scrape = df[df.Read == 0]
    
    # Dictionary to store information
    d1 = {}
    
    # scrape and save information
    for index in df_scrape.index:
        
        # Random Timing
        x = random.randrange(2, 31, 1)
        print(f"Waiting for {x} Seconds.", end='\r')
        time.sleep(x)
        
        # Inits
        d2 = {}
        url = df_scrape.loc[index, 'URL']
        scraper = scrape_me(url)
        
        # save information
        d2['url'] = url
        d2['title'] = scraper.title()
        d2['total_time'] = scraper.total_time()
        d2['yields'] = scraper.yields()
        d2['ingredients'] = scraper.ingredients()
        d2['instructions'] = scraper.instructions()
        d2['rating'] = scraper.ratings()
        d2['id'] = url.split('recipe/')[1].split('/')[0]

        # save
        d1[url] = d2
        df.loc[index, 'Read'] = 1
    
    df.to_csv(path, index=False)
    return d1
```

# My Baking Addiction

```python
# My Baking Addiction
starting = 'https://www.mybakingaddiction.com/cookies/page/'
output = 'recipe_url_database.csv'

# function
df = pd.read_csv(output)
dnew = {}
```

```python
# Obtain links to scrape for recipe urls
index = scrape_me('https://www.mybakingaddiction.com/recipe-index/').links()
starting = []
for link in index[26:]:
    if link['href'] != '/about/':
        starting.append(link['href']+'page/')
    else:
        break
```

```python
# Extract Recipe URLS function
def mybakingaddiction_extract_urls(starting, save_dict):
    for i in range(1, 100):
        try:
            print("Scraping "+starting+str(i))
            page = scrape_me(starting+str(i))
            wait()
            
            # Check if page is blank
            if len(page.links()[26:]) > 43:
                for link in page.links()[26:]:
                    if '/page/' in link['href']:
                        break
                    save_dict[link['href']] = 0
            else:
                break
        except:
            break
    return save_dict
```

```python
# Run through each recipe index page and gather all recipes
for start in starting:
    
    recipe_type = start.split('www.mybakingaddiction.com/')[1].split('/page/')[0]
    dnew[recipe_type] = {}
    dnew[recipe_type] = mybakingaddiction_extract_urls(start, dnew[recipe_type])
```

```python
# Save
with open('my_baking_addiction_scraped.json', 'w') as json_file:
    json.dump(dnew, json_file)
```
