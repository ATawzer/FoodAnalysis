# Creating a recipe dataset
Using regex and titles, query the mongodb to access the relevant recipes. The below example is with Chocolate Chip Cookies.

```python
df = pd.DataFrame(list(recipes.find({"title":{"$regex":"ginger molasses cookie", "$options" :'i'}})))
```

```python
df['parsed_ingredients'] = df.ingredients.apply(parsers.parse_recipe_ingredients)
```

```python
def get_unidentified_ingredients():
    
    pi = []
    ui = []
    for ing in df.parsed_ingredients:
        pi.extend(ing)
    for ing in pi:
        if ing['item'] is None:
            ui.append(ing['ingredient_string'])
    return ui
```
