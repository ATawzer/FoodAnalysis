```python
import pandas as pd
import numpy as np
import json
from tqdm import tqdm_notebook as tqdm
```

```python
ings = json.load(open('train.json'))
```

```python
ing_list = []
for rec in tqdm(ings):
    for ing in rec['ingredients']:
        if ing not in ing_list:
            ing_list.append(ing)
```

```python
# init
ref_dict = {}
for i in tqdm(range(0, len(ing_list))):
        ref_dict[ing_list[i]] = {'types':[]}
```

```python
json.dump(ref_dict, open("ing_library.json", "w"))
```

```python
stopwords = [""]
```

```python
for i in tqdm(range(0, len(ing_list))):
    for j in range(0, len(ing_list)):
        if i != j:
            if " "+ing_list[i]+" " in ing_list[j]:
                ref_dict[ing_list[i]]['types'].append(ing_list[j])
            elif " "+ing_list[j]+" " in ing_list[i]:
                ref_dict[ing_list[i]]['sub_ings'].append(ing_list[j])
```
