```python
import pandas as pd
import json
import pymongo
```

```python
client = pymongo.MongoClient("mongodb://localhost:27017/")
print(client.list_database_names())
```

# Initial Creation Steps

```python
db = client["food_analysis"]
```

```python
urls = db['urls']
```

# Read-in

```python
url_database = json.load(open("url_database.json"))
```

```python
insert_list = []
for source in url_database:
    for url in url_database[source]:
        row = url_database[source][url]
        row['_id'] = url
        insert_list.append(row)
```

```python
urls.insert_many(insert_list)
```

```python
df = pd.DataFrame(list(urls.find({})))
```
